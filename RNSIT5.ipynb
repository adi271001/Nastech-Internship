```python
#Image Processing
from google.colab.patches import cv2_imshow
```

```python
# 1. Read an image
import cv2 #importing the cv2 library
img=cv2.imread('/content/abc.jpg')
cv2_imshow(img)
```

```python
#2. create a duplicate image
import cv2
img=cv2.imread('abc.jpg')
cv2_imshow(img) #displays the image
cv2.imwrite('xyz.png',img)
```

```python
# read info about image
import cv2
img=cv2.imread('abc.jpg') #reads the image
print(img.shape)
print('Height,',int(img.shape[0]),'in pixels')
print('Weight',int(img.shape[1]),'in pixels')
print('Depth',int(img.shape[2]),'in pixels')
# 3 is depth or channel number
# because we use RGB color format in OpenCV
```

```python
#4.GrayScale Image
import cv2
img=cv2.imread('abc.jpg')
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #Converts into grayscale image
cv2_imshow(img) #display original image
cv2_imshow(gray) #display grayscale image
```

```python
#5. Resize an image
import cv2
img=cv2.resize(img,(420,250))
cv2_imshow(img)
```

```python
import cv2
img1=cv2.imread('abc.jpg',0) #read the image
cv2_imshow(img1)
```

```python
#6. Binary Image
import cv2
img=cv2.imread('abc.jpg',0) #0 means to convert it to grayscale
# souce,set value,maximum value,conversion type
_,thresh=cv2.threshold(img,127,255,cv2.THRESH_BINARY)
cv2_imshow(thresh)
```
