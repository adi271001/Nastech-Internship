```python
#SVM - support vector machine
s=['Congrats.You have won lottery and if you want lottery money call the lottery number and get your money','give bank details for lottery money','lottery for sure if bank details are verified']
```

```python
type(s)
```

```python
#tokenization: 
s[1].split()
```

```python
#1. COUNT VECTORIZER
from sklearn.feature_extraction.text import CountVectorizer
vect=CountVectorizer(stop_words='english')
op=vect.fit_transform(s).toarray() #we are converting our sentence into ana array so sklearn understands it
op
```

```python
#stopwords are english words which do nit add much meaning to a sentence
#transformation is done in such a way that the meaning of sentence is not sacrificed
import pandas as pd

data=pd.DataFrame(op,columns=vect.get_feature_names())
data
```

```python
#TFIDF Vectorizer
#tf - term frequency
#idf - inverse document frequency
# how relevant a word is to a sentence
#the values will be in decimal point
from sklearn.feature_extraction.text import TfidfVectorizer
vect=TfidfVectorizer(stop_words='english')
op=vect.fit_transform(s).toarray()
```

```python
data=pd.DataFrame(op,columns=vect.get_feature_names())
data
#it tells us which word is heavier and less heavier in sentence formation
```
